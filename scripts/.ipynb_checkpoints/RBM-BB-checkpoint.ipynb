# RBM BG (todo)

Bernoulli - Gaussian {-1,1}

\begin{align}
q_{v} (s_i) &= \frac{1}{2} (\delta_{s_i,-1} + \delta_{s_i,1}) \\
q_{h} (\tau_i) &= \frac{1}{\sqrt{2\pi \sigma_{h}^{2}}} \exp \left( - \frac{\tau_{a}^{2}}{2\sigma_{h}^2} \right)
\end{align}

\begin{align}
p(\tau | s) \propto \prod_{a} \exp \left( -\frac{\tau_{a}^2}{2\sigma_{h}^2} + \tau_a \sum_{i} w_{ia} s_i \right)
\end{align}

## Define RBM

```python
def _positive_sigmoid(x):
    return 1. / (1. + np.exp(-x))

def _negative_sigmoid(x):
    exp = np.exp(x)
    return exp / (exp + 1.)

def sigmoid(x):
    positive = x >=0
    negative = ~positive
    
    result = np.empty_like(x, dtype=np.float64)
    result[positive] = _positive_sigmoid(x[positive])
    result[negative] = _negative_sigmoid(x[negative])
    
    return result
```

```python
class RBM():
    """
    Binomial Restricted Boltzman Machine
    """
    
    v, h = None, None
    pv, ph = None, None
    history = {'loss':[]}
    
    def __init__(self, n_v, n_h):
        self.n_v = n_v
        self.n_h = n_h
        self._initialize_weights(n_v,n_h)
        
    def _initialize_weights(self, n_v, n_h):
        self.w = np.random.normal(scale=0.01,size=(n_v,n_h)) # weight
#         self.a = np.random.normal(size=(n_v))
#         self.b = np.random.normal(size=(n_h))

        self.a = np.zeros(n_v)
#         self.b = np.zeros(n_h)

#         self.a = np.log(p_i/(1-p_i))
        self.b = -4.0 * np.ones(n_h)
        
    def fit(self, x, lr, epoch):
        
        N = 1.*x.shape[0]
        
        for e in range(epoch):
            
            # data expectation value of vh
            self.ph = sigmoid(self.b + x@self.w)
            self.h = self._sample(self.ph)
            
            v_data = np.sum(x,axis=0)
            h_data = np.sum(self.ph, axis=0)
#             h_data = np.sum(self.h, axis=0)
            vh_data = x.T@self.ph
#             vh_data = x.T@self.h # this is mathematically more accurate but using ph has less sampling noise and slightly faster.

            
            # model expectatoin value of vh
            # CDk algorithm
            self.pv, self.v, self.ph, self.h = self._CDk(1,x)
            
            v_model = np.sum(self.pv, axis=0)
            h_model = np.sum(self.ph, axis=0)
            vh_model = self.pv.T @ self.ph # use probabilities because we don't want random reconstruction.

#             print('vh_model',vh_model)
#             print('v_model',v_model)
#             print('h_model',h_model)
            
            self.w += lr * (vh_data - vh_model)/N
            self.a += lr * (v_data - v_model)/N
            self.b += lr * (h_data - h_model)/N
            
#             print('w_grad',(vh_data - vh_model))
            
            error = np.mean((x - self.v)**2)
            print(e,error)
            self.history['loss'].append(error)
            
    def _CDk(self,k,x):
        pv_n = np.zeros((k,)+x.shape)
        pv_n[0] = sigmoid(self.a + self.h@self.w.T)
        v_n = np.zeros((k,)+x.shape) # (k,n,n_v)
        v_n[0] = self._sample(pv_n[0])
        
        ph_n = np.zeros((k,)+self.ph.shape)
        ph_n[0] = sigmoid(self.b + v_n[0]@self.w)
        h_n = np.zeros((k,)+self.h.shape) # (k,n,n_h)
        h_n[0] = self._sample(ph_n[0])
        
        for i in range(1,k):
            pv_n[i] = sigmoid(self.a + h_n[i-1]@self.w.T)
            v_n[i] = self._sample(pv_n[i])
            
            ph_n[i] = sigmoid(self.b + v_n[i]@self.w)
            h_n[i] = self._sample(ph_n[i])
            
        pv = np.mean(pv_n, axis=0)
        v = np.mean(v_n, axis=0)
        ph = np.mean(ph_n, axis=0)
        h = np.mean(h_n, axis=0)
        
#         print('pv',pv)
        return pv, v, ph, h
            
    def _sample(self, x):
        # {-1,1}
        return 2*np.random.binomial(1,x) - 1
    
    def reconstruct(self,x):
        ph = sigmoid(self.b + x@self.w)
        h = self._sample(ph)
        
        pv = sigmoid(self.a + h@self.w.T)
        v = self._sample(pv)
        # Use probability because we don't want random reconstruction.
        return pv, ph
```
